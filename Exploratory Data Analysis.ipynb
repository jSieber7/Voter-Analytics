# Part 1: POLVIEWS

```python
import pandas as pd
```

```python
df_strings = pd.read_csv('data/data_strings.csv', low_memory=False)
```

```python
order_by = ["Extrmly conservative","Conservative","Slghtly conservative",
                      "Moderate","Slightly liberal","Liberal","Extremely liberal"]
```

```python
#  df.reindex(["Z", "C",
```

```python
ct = pd.crosstab(df_strings.POLVIEWS, df_strings.WRKGOVT)
ct.reindex(order_by)
```

```python
ct = pd.crosstab(df_strings.POLVIEWS, df_strings.WRKGOVT).apply(lambda x: x/x.sum(), axis=1)
ct.reindex(order_by)
```

```python
top_selections = df_strings.groupby('MAJOR1').size().nlargest(10).index

ct = pd.crosstab(df_strings.MAJOR1,df_strings.POLVIEWS).apply(lambda x: x/x.sum(), axis=1)
ct = ct.reindex(top_selections)
ct[order_by]
```
